```python
import warnings

warnings.filterwarnings(action='ignore')
import requests
from bs4 import BeautifulSoup
from datetime import datetime as dt
```

```python
print(dt.now())
```

```python
target_site = 'https://music.bugs.co.kr/chart'
req = requests.get(target_site)
print(req)
html = req.text
# print(html)
soup = BeautifulSoup(html, 'html.parser')
# print(soup)
```

```python
# 노래 제목 크롤링
titles = soup.findAll('p', {'class': 'title'})

for i in range(len(titles)):
    # print(f'{i + 1:3d}위 {titles[i].text.strip()}')
    # print('{:3d}위 {}'.format(i + 1, titles[i].text.split('\n')[1]))
    print('{:3d}위 {}'.format(i + 1, titles[i].text.strip().split('\n')[0]))
```

```python
# 아티스트 크롤링
artists = soup.findAll('p', {'class': 'artist'})

for i in range(len(artists)):
    print('{:3d}위 {}'.format(i + 1, artists[i].text.strip().split('\n')[0]))
```

```python
print(dt.now())
for i in range(100):
    artist = artists[i].text.strip().split('\n')[0]
    title = titles[i].text.strip().split('\n')[0]
    print('{:3d}위 {} - {}'.format(i + 1, artist, title))
```

```python
# 크롤링한 결과를 텍스트 파일에 저장한다.
with open('output/bugsTOP100.txt', 'w', -1, 'UTF-8') as file:
    file.write(dt.now().strftime('%Y년 %m월 %d일 %H시 BUGS 실시간 차트') + '\n' * 2)
    for i in range(100):
        artist = artists[i].text.strip().split('\n')[0]
        title = titles[i].text.strip().split('\n')[0]
        file.write('{:3d}위 {} - {}\n'.format(i + 1, artist, title))
print('bugsTOP100.txt 파일 쓰기 완료!')
```

```python
# 폴더에 저장된 파일 이름 가져오기
import os
fileList = os.listdir('output/')
print(fileList)
print(type(fileList))
for file in fileList:
    # print(file.rstrip('.txt'))
    print(file[:file.rindex('.')])
```

```python
# 텍스트 파일에 저장된 데이터를 읽어서 화면에 출력한다.
try:
    with open('output/bugsTOP100.txt', 'r') as file:
        lines = file.readlines()
        print('bugsTOP100.txt 파일에서 읽기 완료!\n')
        for line in lines:
            print(line.rstrip())
except FileNotFoundError:
    print('디스크에 파일이 존재하지 않습니다.')
```
