```python
import warnings

warnings.filterwarnings(action='ignore')
import requests
from bs4 import BeautifulSoup
from datetime import datetime as dt
```

```python
print(dt.now())
```

```python
target_site = 'https://www.melon.com/chart/index.htm'
header = {'User-agent': 'Mozilla/5.0 (Windows NT 6.3; WOW64; Trident/7.0; rv:11.0) like Gecko'}
req = requests.get(target_site, headers=header)
print(req)  # Response [406]: 허용되지 않음, 요청한 페이지가 요청한 콘텐츠 특성으로 응답할 수 없다.
html = req.text
# print(html)
soup = BeautifulSoup(html, 'html.parser')
# print(soup)
```

```python
# 노래 제목 크롤링
titles = soup.findAll('div', {'class': 'ellipsis rank01'})
for i in range(len(titles)):
    print('{:3d}위 {}'.format(i + 1, titles[i].text.strip()))
```

```python
# 아티스트 크롤링
artists = soup.findAll('span', {'class': 'checkEllipsis'})
for i in range(len(artists)):
    print('{:3d}위 {}'.format(i + 1, artists[i].text))
```

```python
print(dt.now().strftime('%Y년 %m월 %d일 %H시 Melon 차트'))
for i in range(100):
    artist = artists[i].text
    title = titles[i].text.strip()
    print(f'{i+1:3d}위 {artist} - {title}')
```

```python
# 크롤링한 결과를 텍스트 파일에 저장한다.
file = open('output/melonTOP100.txt', 'w')
file.write(dt.now().strftime('%Y년 %m월 %d일 %H시 Melon 차트\n\n'))
for i in range(100):
    artist = artists[i].text
    title = titles[i].text.strip()
    file.write('{:3d}위 {} - {}\n'.format(i + 1, artist, title))
file.close()
print('\nmelonTOP100.txt 파일 쓰기 완료!')
```

```python
# 텍스트 파일에 저장된 데이터를 읽어서 화면에 출력한다.
try:
    file = open('output/melonTOP100.txt', 'r')
    lines = file.readlines()
    print('melonTOP100.txt 파일 읽기 완료!\n')
    for line in lines:
        print(line.strip())
except FileNotFoundError:
    print('디스크에 파일이 존재하지 않습니다.')
```
