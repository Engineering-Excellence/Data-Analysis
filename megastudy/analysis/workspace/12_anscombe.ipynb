```python
import matplotlib as mpl
import seaborn as sns

mpl.rcParams['axes.unicode_minus'] = False
from matplotlib import pyplot as plt

plt.rcParams['font.family'] = 'NanumGothicCoding'
plt.rcParams['font.size'] = 10.0
import warnings

warnings.filterwarnings('ignore')
```

사람의 눈은 수백 줄의 텍스트만으로 이루어진 데이터를 읽거나 기초 통계 수치를 계산하는 방법으로는 데이터를 제대로 분석할 수 없다.
데이터의 숨겨진 패턴을 파악하기 위해서 데이터 시각화를 사용한다.

앤스콤 4분할 그래프
데이터를 시각화하지 않고 수치만 확인했을 때 발생할 수 있는 문제를 보여주기 위한 그래프
앤스콤 데이터  셋은 4개의 그룹으로 구성되어 있으며 4개의 데이터 그룹은 각각 평균, 분산과 같은 수치나 상관관계 회귀선이 모두 같다.
4개의 데이터 셋의 데이터가 모두 같을 것이라고 착각할 수 있다.

```python
# 앤스콤 데이터셋은 seaborn 라이브러리에 포함되어 있다.
# seaborn 라이브러리의 load_dataset 메서드의 인수로 'anscombe'을 전달하면 앤스콤 데이터셋을 불러올 수 있다.
anscombe = sns.load_dataset('anscombe')
anscombe
```

```python
anscombe[anscombe['dataset'] == 'I'].mean()
```

```python
dataset1 = anscombe[anscombe['dataset'] == 'I']
plt.figure(figsize=(8, 6))
plt.plot(dataset1['x'], dataset1['y'], 'o')
plt.show()
```

```python
dataset2 = anscombe[anscombe['dataset'] == 'II']
plt.figure(figsize=(8, 6))
plt.plot(dataset2['x'], dataset2['y'], 'o')
plt.show()
```

```python
dataset3 = anscombe[anscombe['dataset'] == 'III']
plt.figure(figsize=(8, 6))
plt.plot(dataset3['x'], dataset3['y'], 'o')
plt.show()
```

```python
dataset4 = anscombe[anscombe['dataset'] == 'IV']
plt.figure(figsize=(8, 6))
plt.plot(dataset4['x'], dataset4['y'], 'o')
plt.show()
```

```python
# figure() 메서드로 전체 그래프가 그려질 기본 틀을 만든다.
fig = plt.figure()
plt.rcParams['figure.figsize'] = (8, 6)
# add_subplot() 메서드로 그래프 격자를 만든다.
axes1 = fig.add_subplot(2, 2, 1)
axes2 = fig.add_subplot(2, 2, 2)
axes3 = fig.add_subplot(2, 2, 3)
axes4 = fig.add_subplot(2, 2, 4)
# plot() 메서드로 각 격자에 데이터를 전달해서 그래프를 그린다.
axes1.plot(dataset1['x'], dataset1['y'], 'ro')
axes2.plot(dataset2['x'], dataset2['y'], 'go')
axes3.plot(dataset3['x'], dataset3['y'], 'bo')
axes4.plot(dataset4['x'], dataset4['y'], 'yo')
# set_title() 메서드로
axes1.set_title('Dataset I')
axes2.set_title('Dataset II')
axes3.set_title('Dataset III')
axes4.set_title('Dataset IV')
# suptitle() 메서드로 그래프 전체 제목을 추가할 수 있다.
fig.suptitle('Anscombe 4분할 차트')
# 그래프의 제목과 숫자가 겹쳐보이면 tight_layout() 메서드를 실행하면 그래프 레이아웃이 겹치지 않게 조절된다.
fig.tight_layout()
plt.show()
```
