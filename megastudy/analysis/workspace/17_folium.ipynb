```python
# folium은 Leaflet.js를 이용해서 위치 좌표계 값을 지도 위에 시각화 할 수 있다.
import folium
```

# 지도 스타일
Map() 메서드의 tiles 속성으로 지도 스타일을 지정할 수 있다.
- Open Street Map(기본값)
- Stamen Terrain
- Stamen Toner

```python
# folium의 Map() 메서드의 location 속성으로 지정한 위치를 중심으로 하는 zoom_start 속성으로 지정한 배율의 지도를 만든다.
# folium.Map(location=[위도, 경도], zoomstart=배율, tiles='지도 스타일')
folium_map = folium.Map(location=[37.5697987, 126.9837167], zoom_start=18, tiles='Open Street Map')
# folium 지도 객체에서 save() 메서드로 작성된 지도 차트를 html 파일로 저장할 수 있다.
folium_map.save('./output/map.html')
folium_map
```

```python
folium_map = folium.Map(location=[37.5697987, 126.9837167], zoom_start=18, tiles='Stamen Toner')
folium_map.save('./output/map1.html')
folium_map
```

```python
folium_map = folium.Map(location=[37.5697987, 126.9837167], zoom_start=14)

# folium의 Marker() 메서드의 location 속성으로 지정한 위치에 마크를 만든다. → add_to() 메서드로 지도 객체에 추가한다.
# popup 속성으로 마커를 출력했을 때 표시할 문자열을 지정할 수 있다.
folium.Marker(location=[37.579617, 126.977041], popup='경복궁').add_to(folium_map)
folium.Marker(location=[37.5799002, 126.991174], popup='창덕궁').add_to(folium_map)
folium.Marker(location=[37.5658049, 126.9751461], popup='덕수궁').add_to(folium_map)
folium.Marker(location=[37.5787708, 126.9948698], popup='창경궁').add_to(folium_map)

# Popup() 메서드의 max_width 속성으로 마커의 팝업 메시지 크기를 조정할 수 있다.
popup = folium.Popup('영풍문고 - 종로본점', max_width=200)
folium.Marker(location=[37.570176, 126.983197], popup=popup).add_to(folium_map)

folium_map.save('./output/map2.html')
folium_map
```

```python
# Marker() 메서드의 icon 속성에 Icon() 메서드로 마커의 색상과 아이콘 모양을 지정할 수 있다.
# Marker 위에 표시되는 아이콘은 https://getbootstrap.com/docs/3.3/components/#glyphicons-glyph에서 제공하는 아이콘을 표시할 수 있다.

folium_map = folium.Map(location=[37.5697987, 126.9837167], zoom_start=14)

folium.Marker(location=[37.579617, 126.977041], popup='경복궁', icon=folium.Icon(color='red', icon='save')).add_to(folium_map)
folium.Marker(location=[37.5799002, 126.991174], popup='창덕궁', icon=folium.Icon(color='green', icon='plus')).add_to(folium_map)
folium.Marker(location=[37.5658049, 126.9751461], popup='덕수궁', icon=folium.Icon(color='pink', icon='heart')).add_to(folium_map)
folium.Marker(location=[37.5787708, 126.9948698], popup='창경궁', icon=folium.Icon(color='blue', icon='ok')).add_to(folium_map)

popup = folium.Popup('영풍문고 - 종로본점', max_width=200)
folium.Marker(location=[37.570176, 126.983197], popup=popup).add_to(folium_map)

folium_map.save('./output/map3.html')
folium_map
```

```python
folium_map = folium.Map(location=[37.5697987, 126.9837167], zoom_start=14)

# folium의 CircleMarker() 메서드로 원형 마커를 만들 수 있다.
# radius 속성으로 지도에 표시할 원의 반지름을 지정한다.
# color 속성으로 원의 색상, fill_color 속성으로 원의 내부 색상을 지정한다.

folium.CircleMarker(location=[37.579617, 126.977041], popup='경복궁', radius=80, color='red', fill_color='blue').add_to(folium_map)
folium.CircleMarker(location=[37.5799002, 126.991174], popup='창덕궁', radius=30, color='#000000', fill_color='#pink').add_to(folium_map)
folium.CircleMarker(location=[37.5658049, 126.9751461], popup='덕수궁').add_to(folium_map)
folium.CircleMarker(location=[37.5787708, 126.9948698], popup='창경궁').add_to(folium_map)

popup = folium.Popup('영풍문고 - 종로본점', max_width=200)
folium.CircleMarker(location=[37.570176, 126.983197], popup=popup, radius=50, color='red', fill_color='blue').add_to(folium_map)

folium_map.save('./output/map4.html')
folium_map
```
