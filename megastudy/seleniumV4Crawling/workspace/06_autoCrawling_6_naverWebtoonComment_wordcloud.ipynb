```python
import warnings

warnings.filterwarnings(action='ignore')
import numpy as np
import matplotlib as mpl

mpl.rcParams['axes.unicode_minus'] = False
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = 'NanumGothicCoding'
plt.rcParams['font.size'] = 15
```

```python
from wordcloud import WordCloud
from PIL import Image
import nltk
from konlpy.tag import Twitter

t = Twitter()
```

```python
# 네이버 웹툰 리뷰가 저장된 파일을 읽는다.
review = open('./review.txt', encoding='UTF-8').read()
review
```

```python
# 한글 형태소 분석을 실행한다.
review_tokens = t.nouns(review)
review_tokens
```

```python
# 불용어를 정의한다.
stop_words = ['달라', '그거', '임펜', '이번', '뭔가', '뿜빠이', '캬캬캭', '진짜', '뿌엉']
review_tokens = [each_word for each_word in review_tokens if each_word not in stop_words and len(each_word) > 1]
review_tokens
```

```python
# 워드클라우드로 구현할 단어들의 출현 빈도수를 센다.
token = nltk.Text(review_tokens)
token.vocab().most_common(200)
```

```python
# 워드클라우드로 구현할 단어를 뽑아내서 단어를 key로 하고 출현 빈도수를 value로 하는 딕셔너리를 만든다.
data = dict(token.vocab().most_common(200))
data
```

```python
# 워드클라우드 마스크 이미지
mask = np.array(Image.open('./mask.png'))
```

```python
# 딕셔너리에 저장된 데이터는 generate_from_frequencies() 메서드를 사용해서 워드클라우드를 만든다.
wordcloud = WordCloud(font_path='/Users/kyle/Library/Fonts/NanumGothicCoding.ttf', background_color='white',
                      max_font_size=200, max_words=200, mask=mask).generate_from_frequencies(data)
plt.figure(figsize=(12, 8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
