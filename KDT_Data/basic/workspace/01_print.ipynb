## '%'와 서식 문자를 사용해서 출력하기
### 서식 문자
d(정수), f(실수), s(문자열)을 사용해서 출력 서식을 지정할 수 있고, 출력 서식을 제외한 나머지 문자들은 그대로 출력된다.
### 출력 서식의 형식
%[-][0][n][.m]서식문자 → [] 안의 내용은 생략할 수 있다.
n: 전체 자릿수, 전체 자릿수가 지정되면 데이터는 오른쪽으로 정렬되어 출력된다.
-: 전체 자릿수가 지정된 경우 데이터를 왼쪽으로 정렬해서 출력한다.
0: 전체 자릿수가 지정된 경우 숫자 왼쪽 남는 자리에 '0'을 채워서 출력한다. 문자열일 경우 무시된다.
.m: 실수를 출력할 경우 소수점 아래 자릿수를 지정한다.

```python
print(100)
print(1)
print(10000)
```

print('출력 서식' %출력할 데이터) → 출력할 데이터가 1개일 경우

```python
print('%5d' % 100)
print('%5d' % 1)
print('%-5d' % 1)
# 파이썬은 1줄에 1문장 코딩을 원칙으로 하고 부득이하게 1줄에 2문장 이상을 코딩해야 할 경우 문장과 문장 사이를 ';'으로 구분한다.
print('%05d' % 1);  print('%5d' % 10000)
```

```python
print('abc')
print('a')
print('abcde')
```

```python
print('%5s' % 'abc')
print('%5s' % 'a')
print('%-5s' % 'a')
print('%05s' % 'a')  # C/C++/C#는 0이 출력되고, Java는 에러가 발생되며, Python은 무시한다.
print('%5s' % 'abcde')
```

```python
print('%6.2f' % 123.45)
print('%6.2f' % 123.456)  # 소수점 아래 숫자는 잘리는 자리에서 반올림하여 출력한다. → 실제 데이터는 변하지 않는다!
print('%6.2f' % 123.4)  # 남는 자리는 0을 채워서 출력한다.
```

'%'를 사용한 출력 서식에 맞춰 출력할 데이터가 2건 이상이라면 출력할 데이터를 반드시 ()로 묶어야 한다.

```python
print('%d + %d = %d' % (5, 3, 5 + 3))  # 덧셈
print('%d + %d = %d' % (5, 3, 5 - 3))  # 뺄셈
print('%d + %d = %d' % (5, 3, 5 * 3))  # 곱셈

# 파이썬은 '/' 연산자로 나눗셈을 하면 소수점 아래까지 계산된다.
# '/' 연산자는 결과가 실수로 나오지만 출력 서식이 '%d'(정수)로 지정되면 정수만 출력된다.
print('%d + %d = %d' % (5, 3, 5 / 3))  # 나눗셈 → 결과가 실수
print('%d + %d = %f' % (5, 3, 5 / 3))  # 나눗셈
print('%d + %d = %d' % (5, 3, 5 // 3))  # 나눗셈 → 결과가 정수
print('%d + %d = %f' % (5, 3, 5 // 3))  # 나눗셈

# 출력 서식을 지정할 때 '%' 문자 자체를 출력하려면 연속해서 2개를 써야 한다.
print('%d %% %d = %d' % (5, 3, 5 % 3))  # 나머지
print('%d의 %d제곱 = %d' % (2, 10, 2 ** 10))  # 거듭제곱
print('제곱근 %d = %f' % (2, 2 ** 0.5))  # 제곱근
```

# 산술 연산자 우선순위
() → ** → *, /, //, % → +, -
'%'로 서식을 지정할 때는 %d, %f, %s 위치에 데이터가 대입되어 출력되었지만, {}를 사용하면 {}에 데이터가 대입되어 출력된다.

```python
print('{} + {} = {}'.format(5, 3, 5 + 3))
print('{} - {} = {}'.format(5, 3, 5 - 3))
print('{} * {} = {}'.format(5, 3, 5 * 3))
print('{} / {} = {}'.format(5, 3, 5 / 3))
print('{} // {} = {}'.format(5, 3, 5 // 3))
print('{} % {} = {}'.format(5, 3, 5 % 3))
print('{}의 {}제곱 = {}'.format(2, 10, 2 ** 10))
print('제곱근 {} = {}'.format(2, 2 ** 0.5))
```

format() 메서드와 출력할 데이터의 인덱스(일련번호)를 사용해서 출력하기 → 인덱스는 0부터 시작된다.

```python
print('{} {}'.format('홍길동', '바보'))

# 인덱스를 지정하지 않은 상태에서 아래와 같이 {}가 출력할 데이터보다 많으면 에러가 발생한다!
# print('{} {} {}'.format('홍길동', '바보')) # Error

# {}가 모자라면 에러가 발생되지 않고 출력되며 남는 데이터가 무시된다.
print('{}'.format('홍길동', '바보'))
```

```python
print('{0} {1}'.format('홍길동', '바보'))
print('{1} {0}'.format('홍길동', '바보'))
# {}에 출력할 데이터의 인덱스를 지정하면 데이터와 {}의 개수가 다르더라도 인덱스 범위만 벗어나지 않으면 정상 출력된다.
print('{0} {1} {0}'.format('홍길동', '바보'))

# {}에 지정한 인덱스가 범위를 벗어나면 에러가 발생한다.
# print('{0} {1} {2}'.format('홍길동', '바보'))    # Error

# {}에 인덱스를 부분적으로 지정하면 에러가 발생한다.
# print('{0} {1} {}'.format('홍길동', '바보'))    # Error
```

출력 서식과 format() 메서드로 출력할 데이터의 서식을 지정해서 출력하기
{인덱스:출력서식} → 인덱스는 가능하고 출력 서식에 '%'를 사용하지 않는다.

```python
print('%10s은(는) %7d원입니다.' %('멜론', 15000))
print('{}은(는) {}원입니다.'.format('멜론', 15000))
print('{0}은(는) {1}원입니다.'.format('멜론', 15000))
print('{0:10s}은(는) {1:7d}원입니다.'.format('멜론', 15000))
print('{:10s}은(는) {:7d}원입니다.'.format('멜론', 15000))
print('{:10s}은(는) {:7,d}원입니다.'.format('멜론', 15000)) # 천 단위 구분 기호 넣기
print('{:10s}은(는) {:7.3%}입니다.'.format('물가상승율', 0.142857)) # 백분율, 자동으로 100을 곱해서 표시한다.
```

```python
# '<', '^', '>'를 사용해서 문자열 정렬 방법을 지정할 수 있다. → '<'가 기본값
print('{:10s}은(는) {:7d}원입니다.'.format('멜론', 15000))
print('{:<10s}은(는) {:7d}원입니다.'.format('멜론', 15000))
print('{:^10s}은(는) {:7d}원입니다.'.format('멜론', 15000))
print('{:>10s}은(는) {:7d}원입니다.'.format('멜론', 15000))
```

```python
# 문자열 출력시 남는 자리의 공백을 채울 채움 문자(1글자)를 '<', '^', '>' 앞에 지정할 수 있다.
# print('{:_10s}은(는) {:7d}원입니다.'.format('멜론', 15000)) # Error
print('{:_<10s}은(는) {:7d}원입니다.'.format('멜론', 15000))
print('{:*<10s}은(는) {:7d}원입니다.'.format('멜론', 15000))
print('{:#^10s}은(는) {:7d}원입니다.'.format('멜론', 15000))
print('{:@>10s}은(는) {:7d}원입니다.'.format('멜론', 15000))
```
