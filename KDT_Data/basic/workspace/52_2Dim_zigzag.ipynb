```python
def matrix(data):
    for i in range(len(data)):
        for j in range(len(data[0])):
            print('{:3d}'.format(data[i][j]), end=' ')
        print()
```

```python
row, col = map(int, input('2차원 리스트를 만들 행과 열의 개수: ').split())
data = [[0] * col for i in range(row)]
n = 0  # 리스트에 1씩 증가하며 채워질 숫자가 기억되는 변수
for i in range(len(data)):  # 행
    # 짝수행인지 홀수행인지 판단한다.
    if i % 2 == 0:
        # 짝수행은 왼쪽에서 오른쪽으로 숫자를 채운다.
        for j in range(len(data[i])):
            n += 1
            data[i][j] = n
        # ===== for j
    else:
        # 홀수행은 오른쪽에서 왼쪽으로 숫자를 채운다.
        for j in range(len(data[i]) - 1, -1, -1):
            n += 1
            data[i][j] = n
        # ===== for j
    # ===== if
# ===== for i

matrix(data)
# print(np.array(data))
```

```python
row, col = map(int, input('2차원 리스트를 만들 행과 열의 개수: ').split())
data = [[0] * col for i in range(row)]
n = 0
start = 0   # 2차원 리스트에 숫자가 채워지기 시작하는 열 인덱스
end = col - 1   # 2차원 리스트에 숫자가 채워지는 마지막 열 인덱스
sw = 1  # start부터 end까지 증가값으로 사용할 변수

for i in range(row):
    # i = 0, sw = 1, j = range(0, 5, 1) → 0, 1, 2, 3, 4
    # i = 1, sw = -1, j = range(4, -1, -1) → 4, 3, 2, 1, 0
    # i = 2, sw = 1, j = range(0, 5, 1) → 0, 1, 2, 3, 4
    # i = 3, sw = -1, j = range(4, -1, -1) → 4, 3, 2, 1, 0
    for j in range(start, end + sw, sw):
        n += 1
        data[i][j] = n
    start, end = end, start
    sw *= -1

matrix(data)
```

```python
row, col = map(int, input('2차원 리스트를 만들 행과 열의 개수: ').split())
data = [[0] * col for i in range(row)]
n = 0
start = [i for i in range(col)] # 왼쪽부터 오른쪽으로 진행하는 열 인덱스를 기억하는 리스트를 선언한다.
end = [i for i in range(col - 1, -1, -1)]   # 오른쪽부터 왼쪽으로 진행하는 열 인덱스를 기억하는 리스트를 선언한다.

for i in range(row):
    for j in start:
        n += 1
        data[i][j] = n
    # ===== for j
    start, end = end, start
# ===== for i

matrix(data)
```
