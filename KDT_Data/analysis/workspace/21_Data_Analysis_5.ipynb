# to_datetime()
object → datetime

```python
# 집행일시 열을 이용해서 '연', '월', '일', '연월', '시', '분', '요일' 파생 변수를 만든다.
df['집행일시'] = pd.to_datetime(df['집행일시'])
df.dtypes
```

```python
print(type(df['집행일시']))
print(type(df['집행일시'].dt))  # SR에 datetime 메서드를 적용할 경우 dt로 1회성 변환 필요
print(type(df['집행일시'].dt.year))
# print(df['집행일시'].dt.year)     # int64(8바이트, long)메모리 과다 차지
df['집행일시'].dt.year.astype(np.int32)  # int32(4바이트, int)
```

```python
df['연'] = df['집행일시'].dt.year.astype(np.int32)
df['월'] = df['집행일시'].dt.month.astype(np.int32)
df['일'] = df['집행일시'].dt.day.astype(np.int32)
df['연월'] = df['연'].astype(str) + '-' + df['월'].astype(str)
df['시'] = df['집행일시'].dt.hour.astype(np.int32)
df['분'] = df['집행일시'].dt.minute.astype(np.int32)
# 0: 월, 1: 화, ... 6: 일
df['요일'] = df['집행일시'].dt.dayofweek
print(df.dtypes)
df
```

```python
# 집행장소의 주소는 전처리를 통해 제거하고 상호명이라는 컬럼을 만들어 집행장소에서 상호명을 가져온다.
# print(df['집행장소'].str.split('(')[0][0].strip())
# for _, data in df.iterrows():
temp = df['집행장소'].apply(lambda x: x.split('(')[0].strip())
df['상호명'] = temp.apply(lambda x: x.split(' ')[0].strip())
df
```

```python
# 상호명이 '대상자'인 데이터의 집행목적을 확인한다.
df.loc[df['상호명'] == '대상자', '집행목적'].value_counts().head()
```
