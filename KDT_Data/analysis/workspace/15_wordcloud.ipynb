워드클라우드를 그레이톤으로 표시하는 함수를 만든다.
- word: 워드클라우드로 표시할 단어
- font_size: 워드클라우드로 표시할 단어의 크기
- position: 워드클라우드로 표시할 단어의 위치
- orientation: 워드클라우드로 표시할 단어의 방향
- random_state: 워드클라우드로 표시되는 단어들의 위치 seed
- **hwargs: 워드클라우드로 표시할 단어의 글꼴이 저장된 경로(*: tuple, **: dict)

```python
def gray_color_func(word, font_size, position, orientation, random_state=None, **hwargs):
    # print(word, font_size, position, orientation, random_state, hwargs)
    return 'hsl(9, 0%%, %d%%)' % random.randint(0, 60)
```

```python
plt.figure(figsize=(10, 8))
# 워드클라우드를 랜덤한 그레이스케일로 표시하려면 recolor() 메서드의 color_func 속성으로 그레이스케일을 리턴하는 메서들르 실행한다.
plt.imshow(wordcloud.recolor(color_func=gray_color_func, random_state=100), interpolation='bilinear')
plt.axis('off')
plt.show()
```

# 한글 워드클라우드

```python
import nltk
from konlpy.corpus import kolaw
# from konlpy.tag import Twitter
# t = Twitter()
from konlpy.tag import Okt

t = Okt()
```

```python
ko_con_text = kolaw.open('constitution.txt').read()
ko_con_text
```

```python
# 한글 형태소 분석을실행한다.
tokens_ko = t.nouns(ko_con_text)
tokens_ko
```

```python
# 불용어를 정의한다.
stop_words = ['제', '월', '일', '조', '수', '때', '그', '이', '바', '및', '안', '정', '관', '위', '장', '거나']
tokens_ko = [each_word for each_word in tokens_ko if each_word not in stop_words]
```

```python
# 워드클라우드로 구현할 단어들의 출현 빈도수를 센다.
ko = nltk.Text(tokens_ko)
ko.vocab().most_common(50)
```

```python
# 워드클라우드로 구현할 단어를 추출해서 단어를 key로 하고 출현 빈도수를 value로 하는 딕셔너리를 만든다.
data = dict(ko.vocab().most_common(500))
data
```

```python
# 딕셔너리에 저장된 데이터는 generate_from_frequencies() 메서드를 사용해서 워드클라우드를 만든다.
wordcloud = WordCloud(font_path='/Users/kyle/Library/Fonts/NanumGothicCoding.ttf', background_color='white',
                      relative_scaling=0.5).generate_from_frequencies(data)
plt.figure(figsize=(10, 8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
